```python
import pandas as pd
```

```python
df = pd.read_csv("retail_sales_practice.csv")
df.head()
```

```python
df.info()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
df = df.dropna(subset=['Sales'])
```

```python
df['Category'] = df['Category'].fillna("Unknown")
```

```python
df['Discount'] = df['Discount'].fillna(0)
```

```python
df[(df['Quantity'] == 0) & (df['Sales'] > 0)].shape[0]
```

```python
df[df['Quantity'] == 0]['Product'].value_counts()
```

```python
df = df[~((df['Quantity'] == 0) & (df['Sales'] > 0))]
```

```python
df[df['Profit'] > df['Sales']].shape[0]
```

```python
df[df['Profit'] > df['Sales']].head()
```

```python
df = df[df['Profit'] <= df['Sales']]
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df[df['Quantity'] <= 0]
```

```python
df[(df['Discount'] < 0) | (df['Discount'] > 1)]
```

```python
df['Discount'] = df['Discount'] / 100
```

```python
df['Discount'].describe()
```

```python
df['City'] = df['City'].str.strip()        # remove spaces
df['City'] = df['City'].str.title()        # make proper case
df['City'] = df['City'].replace('Dehli', 'Delhi')  # fix typo
```

```python
df['City'].unique()
```

```python
df['Category'].unique()
```

```python
df[df['Category'] == 'Unknown'].shape
```

```python
df['Category'] = df['Category'].replace('Unknown', 'Others')
```

```python
df['Category'].value_counts()
```

```python
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
df['Profit_Margin'] = df['Profit'] / df['Sales']
```

```python
df.head()
```

```python
df['Profit_Margin'].describe()
```

```python
df[df['Profit_Margin'] > 0.9]
```

```python
df.to_csv("cleaned_retail_sales.csv", index=False)
```
